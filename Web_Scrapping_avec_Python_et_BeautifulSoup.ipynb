# importer les packages

```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
```

# saisir l'URL

```python
url=' http://feeds.bbci.co.uk/news/rss.xml'
```

# importer le code de la page

```python
reponse=requests.get(url)
Soup=BeautifulSoup(reponse.text,"html.parser")
```

la on a envoyé une requete vers l'url et puis on la mis dans une variable réponse après cette réponse la on a fait passer par la fonction "BeautifulSoup" qu'on a importé de bs4 


```python
Soup
```

```python
items=Soup.findAll('item')
```

```python
items
```

```python
items[0]
```

```python
item=items[0]
```

```python
item.title.text
```

```python
item.description.text
```

```python
new_items=[]
```

la on va créer une nouvelle liste dans laquelle on va mettre des information extraites à partir de la liste précédente

```python
for i in items:# la boucle for pour parcourir tous les items
  news_i={}#création d'un dictionnaire
  news_i['title']=i.title.text
  news_i['description']=i.description.text
  news_i['pubdate']=i.pubdate.text
  new_items.append (news_i) 
```

```python
new_items
```

## convertir la liste en base de donées

```python
df=pd.DataFrame(new_items,columns=['title','description','pubdate'])
```

```python
df.to_csv('web_scrapping.cvs',index=False,encoding='utf-8')
```
